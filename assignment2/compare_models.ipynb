```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# [lstm, cnn, fcnn]
train_losses = [0.014855530554414145, 0.0015098792181061072, 0.0048810603783210425]
val_losses = [0.018546695155860434, 0.006026702174681857, 0.004793478560353082]
test_losses = [0.0803943583207136, 0.007291555298775446, 0.17344030100778116]
```

```python
fig, ax = plt.subplots()
bar_width = 0.25
opacity = 0.8

# Create bar plots
train_bars = ax.bar(np.arange(len(train_losses)), train_losses, bar_width,
                     alpha=opacity, color='b', label='Train')
val_bars = ax.bar(np.arange(len(val_losses)) + bar_width, val_losses, bar_width,
                     alpha=opacity, color='g', label='Validation')
test_bars = ax.bar(np.arange(len(test_losses)) + 2*bar_width, test_losses, bar_width,
                     alpha=opacity, color='r', label='Test')

# Set labels and title
ax.set_xlabel('Model')
ax.set_ylabel('Loss')
ax.set_title('Summary of Training, Validation, and Test Set Errors for Different Models')
ax.set_xticks(np.arange(len(train_losses)) + bar_width)
ax.set_xticklabels(('LSTM', 'CNN', 'FCNN'))
ax.legend()

# Display plot
plt.show()
```
